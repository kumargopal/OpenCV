# OpenCV : Introduction and loading Image

#### Importing essential libraries

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

#### Loading image

```python
image = cv.imread('Watch.jpg', cv.IMREAD_ANYCOLOR)
```

#### Ploting image using OpenCV-imshow

```python
cv.imshow('Watch', image)
cv.waitKey(0)
cv.destroyAllWindows()
```

#### Ploting image using Matplotlib-imshow

```python
plt.imshow(image, cmap = 'gray', interpolation='bicubic')
plt.show()
```

#### Ploting a line in an image using matplotlib-plot

```python
plt.imshow(image, cmap = 'gray', interpolation='bicubic')
plt.plot([120,100], [280,100], 'c', linewidth = 5)
plt.show()
```

#### Reading an image in gray-scale and saving it

```python
grayImage = cv.imread('Watch.jpg', cv.IMREAD_GRAYSCALE)
cv.imwrite('grayWatch.jpg', grayImage)
writtenImage = cv.imread('grayWatch.jpg')
plt.imshow(writtenImage)
plt.show()
```
